```python
model = Sequential()
model.add(Dense(4, input_dim=1, activation='linear', name='layer_1'))
model.add(Dense(1, activation='linear', name='layer_2'))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

# Section 2
# Simple custom loss for nn model using hidden layer weights

```python
x = np.array([np.random.randn(100), np.random.randn(100)])
y = x[0] * 3 + np.random.randn(100) * 0.8

x = x.reshape((100, 2))
```

```python
inputs = keras.Input(shape=(2,))
cls1a = Dense(2, activation='relu')(inputs)
output = Dense(1, activation='sigmoid')(cls1a)
model = keras.Model(inputs=inputs, outputs=output, name="simple_model")
model.summary()
```

```python
print("Weights and biases of the layers after training the model: \n")
for layer in model.layers:
    print(layer.name)
    if layer.get_weights():
        print("Weights")
        print("Shape: ",layer.get_weights()[0].shape,'\n',layer.get_weights()[0])
        print("Bias")
        print("Shape: ",layer.get_weights()[1].shape,'\n',layer.get_weights()[1],'\n')
```

```python
def minimize(y_true, y_pred):
    # goal is to reduce overall weight values
    weights = cls1a
    abs_weights = tf.math.abs(weights)
    reg = tf.math.reduce_sum(abs_weights)
    
    ret = keras.losses.mean_squared_error(y_true, y_pred) + 100 * reg
    
    return ret
```

```python
model.compile(
    loss=minimize,
    optimizer='adam',
    metrics=["accuracy", "mean_absolute_error"]
)

history = model.fit(x, y, batch_size=64, epochs=1000, validation_split=0.2)
```

```python
print("Weights and biases of the layers after training the model: \n")
for layer in model.layers:
    print(layer.name)
    if layer.get_weights():
        print("Weights")
        print("Shape: ",layer.get_weights()[0].shape,'\n',layer.get_weights()[0])
        print("Bias")
        print("Shape: ",layer.get_weights()[1].shape,'\n',layer.get_weights()[1],'\n')
```

# Section 3
# Attribute Orthogonal Regularization

- Define a split model system: cls1 and cls2 are two hidden layers that we will focus on
- The customer loss function, `orthogonal_loss`, uses symbolic link to the layers built through the Keras functional api

```python
x = np.random.randn(10000,2)
y = x[:, 1] + 1
```

```python
inputs = keras.Input(shape=(2,), name='input')

cls1 = Dense(2, name='cls1', activation='relu')(inputs)
cls2 = Dense(2, name='cls2', activation='relu')(inputs)

output1 = Dense(1, name='output1', activation='sigmoid')(cls1)
output2 = Dense(1, name='output2', activation='sigmoid')(cls2)

model = keras.Model(inputs=inputs, outputs=[output1, output2] , name="simple_model")
model.summary()
```

```python
print("Weights and biases of the layers after training the model: \n")
cls1_weights, cls2_weights = None, None
for layer in model.layers:
    print(layer.name)
    if layer.name == 'cls1': cls1_weights = layer.get_weights()[0]
    if layer.name == 'cls2': cls2_weights = layer.get_weights()[0]
    if layer.get_weights():
        print("Weights")
        print("Shape: ",layer.get_weights()[0].shape,'\n',layer.get_weights()[0])
        print("Bias")
        print("Shape: ",layer.get_weights()[1].shape,'\n',layer.get_weights()[1],'\n')
```

### Here, we'll check the correlation of the weights in the hidden layers prior to training.

```python
cls1_cls2 = np.matmul(cls1_weights, cls2_weights.T)
numerator = np.linalg.norm(cls1_cls2, ord=1)
cls1_norm = np.linalg.norm(cls1_weights, ord=2)
cls2_norm = np.linalg.norm(cls2_weights, ord=2)
denominator = cls1_norm * cls2_norm
loss_ortho = np.divide(numerator, denominator + 1e-10)
print(loss_ortho)
```

```python
cls1_weights, cls2_weights
```

```python
def orthogonal_loss(y_true, y_pred):
    # requires cls1, cls2 layers to be defined through functional API
    cls1_cls2 = tf.linalg.matmul(cls1, cls2, transpose_b=True)
    numerator = tf.norm(cls1_cls2, ord=1)
    cls1_norm = tf.norm(cls1, ord=2)
    cls2_norm = tf.norm(cls2, ord=2)
    denominator = tf.math.multiply(cls1_norm, cls2_norm)
    loss_ortho = tf.math.divide_no_nan(numerator, denominator)
    
    loss_categorical = keras.losses.categorical_crossentropy(y_true, y_pred)
    
    return loss_categorical + loss_ortho
```

```python
model.compile(
    loss=orthogonal_loss,
    optimizer='adam',
    metrics=["accuracy", "mean_absolute_error"]
)

history = model.fit(x, [y, y], batch_size=64, epochs=20, validation_split=0.2)
```

```python
print("Weights and biases of the layers after training the model: \n")
cls1_weights, cls2_weights = None, None
for layer in model.layers:
    print(layer.name)
    if layer.name == 'cls1': cls1_weights = layer.get_weights()[0]
    if layer.name == 'cls2': cls2_weights = layer.get_weights()[0]
    if layer.get_weights():
        print("Weights")
        print("Shape: ",layer.get_weights()[0].shape,'\n',layer.get_weights()[0])
        print("Bias")
        print("Shape: ",layer.get_weights()[1].shape,'\n',layer.get_weights()[1],'\n')
```

### Now, we'll check the correlation of the hidden layer parameters after training.

```python
cls1_cls2 = np.matmul(cls1_weights, cls2_weights.T)
numerator = np.linalg.norm(cls1_cls2, ord=1)
cls1_norm = np.linalg.norm(cls1_weights, ord=2)
cls2_norm = np.linalg.norm(cls2_weights, ord=2)
denominator = cls1_norm * cls2_norm
loss_ortho = np.divide(numerator, denominator + 1e-10)
print(loss_ortho)
```
